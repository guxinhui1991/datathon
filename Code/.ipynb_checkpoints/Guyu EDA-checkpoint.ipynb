```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('data/listing_formatted_hr_vol_fac_removed.csv', dtype = {'dtype' : np.float64})
```

```python
out_faci_columns = data.columns[51:]
pure_data = data[['hit_rate'] + out_faci_columns.values.tolist()]
pure_data = pure_data.dropna()
```

```python
y = pure_data['hit_rate']
X = pure_data[out_faci_columns]
```

```python
corr_list = []
for col_name in out_faci_columns:
    corr_list.append(np.corrcoef(y.values, X[col_name].values)[0][1])
```

```python
corr_data = pd.DataFrame({'facility name' : out_faci_columns, 'correlation': corr_list}).sort_values(by = 'correlation', ascending = False)
```

```python
fig = plt.figure(figsize = (12, 20))
ax = fig.add_subplot(111)
width = 0.35
ind = np.arange(corr_data.shape[0])
a = ax.barh(ind, corr_data['correlation'], width)
ax.set_yticks(ind + width / 2)
ax.set_yticklabels(corr_data['facility name'])
```
