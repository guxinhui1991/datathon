```python
venue_places = venue_raw[['longitude', 'latitude', 'types']]
```

```python
type_list = []
all_faci_list = []
for index, row in venue_places.iterrows():
    s = row['types'].lstrip('[')
    s = s.rstrip(']')
    sl = s.split(',')
    tmp = []
    for s in sl:
        s = s.lstrip('\'').rstrip('\'').replace('\'', '').strip()
        tmp.append(s)
        if s in type_list:
            continue
        type_list.append(s)
    all_faci_list.append(tmp)

full_list = ['latitude', 'longitude'] + type_list
```

```python
type_list
```

```python
out_faci = pd.DataFrame(np.zeros((len(venue_places), len(full_list))), columns = full_list)

for index, row in venue_places.iterrows():
    out_faci.iloc[index]['longitude'] = row['longitude']
    out_faci.iloc[index]['latitude'] = row['latitude']
    out_faci.iloc[index][all_faci_list[index]] = 1
                     
out_faci
```

```python
listing_pure_ol = listing_pure[listing_pure.city == 'oakland']
```

```python
out_faci_ol = out_faci[out_faci.city == 'oakland']
```

```python
listing_pure_ol.head()
```

```python
out_faci_ol.head()
```

```python
threshold = 0.0000135215   # 4 blocks distance
```

```python
listing_pure_ol_val = listing_pure_ol.values
out_faci_ol_val = out_faci_ol.values
res = []
for i in np.arange(listing_pure_ol_val.shape[0]):
    lati = listing_pure_ol_val[i][1]
    longi = listing_pure_ol_val[i][2]
    tmp = []
    for j in np.arange(out_faci_ol_val.shape[0]):
        dist = (lati - out_faci_ol_val[j][0])**2 + (longi - out_faci_ol_val[j][1])**2
        if dist < threshold:
            tmp.append(out_faci_ol_val[j][2:-1])
    tmp = np.array(tmp)
    i_res = tmp.sum(axis = 0) > 0
    i_res = np.append(np.array(listing_pure_ol_val[i][0]), i_res)
    res.append(i_res)
```

```python
col_name = ['id'] + type_list
final_res = pd.DataFrame(res, columns = col_name)
final_res
```
