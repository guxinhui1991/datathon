```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
listing_raw = pd.read_csv('data/listings.csv', dtype = {'latitude': np.float64, 'longitude': np.float64})
```

```python
venue_raw = pd.read_csv('data/venues.csv',dtype = {'latitude': np.float64, 'longitude': np.float64})
```

```python
# convert the venue table to a more meaningful representation
venue_places = venue_raw[['longitude', 'latitude', 'types']]
type_list = []
all_faci_list = []
for index, row in venue_places.iterrows():
    s = row['types'].lstrip('[')
    s = s.rstrip(']')
    sl = s.split(',')
    tmp = []
    for s in sl:
        s = s.lstrip('\'').rstrip('\'').replace('\'', '').strip()
        tmp.append(s)
        if s in type_list:
            continue
        type_list.append(s)
    all_faci_list.append(tmp)
```

```python
# check each location the facility type availability
full_list = ['latitude', 'longitude'] + type_list
out_faci = pd.DataFrame(np.zeros((len(venue_places), len(full_list))), columns = full_list)

for index, row in venue_places.iterrows():
    out_faci.iloc[index]['longitude'] = row['longitude']
    out_faci.iloc[index]['latitude'] = row['latitude']
    out_faci.iloc[index][all_faci_list[index]] = 1
                     
out_faci
```

```python
out_faci['city'] = venue_raw['city'].values
```

```python
# get the listing's info, mainly city, longtitude and latitude and id
listing_pure = listing_raw[['id', 'latitude', 'longitude', 'city']]
```

```python
# calculate the near facility of each listing
def get_out_facility_info(city):
    listing_pure_city = listing_pure[listing_pure.city == city]
    out_faci_city = out_faci[out_faci.city == city]
    
    threshold = 0.0000135215   # 4 blocks distance
    
    listing_pure_city_val = listing_pure_city.values
    out_faci_city_val = out_faci_city.values
    res = []
    for i in np.arange(listing_pure_city_val.shape[0]):
        lati = listing_pure_city_val[i][1]
        longi = listing_pure_city_val[i][2]
        tmp = []
        for j in np.arange(out_faci_city_val.shape[0]):
            dist = (lati - out_faci_city_val[j][0])**2 + (longi - out_faci_city_val[j][1])**2
            if dist < threshold:
                tmp.append(out_faci_city_val[j][2:-1])
        tmp = np.array(tmp)
        i_res = tmp.sum(axis = 0) > 0
        i_res = np.append(np.array(listing_pure_city_val[i][0]), i_res)
        res.append(i_res)
    col_name = ['id'] + type_list
    final_res = pd.DataFrame(res, columns = col_name)
        
    return final_res
```

```python
oakland_faci = get_out_facility_info('oakland')
```
